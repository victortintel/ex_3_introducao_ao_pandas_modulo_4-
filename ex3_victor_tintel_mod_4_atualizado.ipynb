````python
#1) Seu código aqui
def qual_url(uf, interesse):
    url=(f"https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={uf}&tipo_interessado={interesse}")
    r = requests.get(url)
    r.status_code
    data_json = r.json()
    tabela = pd.DataFrame(data_json["items"])
    return tabela
````

````python
qual_url("MG", "Estado")
````
